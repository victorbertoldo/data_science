```python
import pandas as pd
```

```python
dados = pd.read_csv('out/aluguel_residencial_notna.csv', sep=';')
dados.head()
```

### Adicionando a variável  ->  `Valor Bruto`

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head()
```

### Adicionando a variável  ->  `Valor m²`

```python
dados['Valor m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head()
```

### Precisamos agora definir os imóveis que são algum tipo de casa (moradia) como apenas 'Casa' e assim tbm os que são apartamento, afim de agregarmos as informações para criar uma analise mais geral
#### Para isso iremos utilizar função lambda e para tal precisamos criar uma lista 

```python
dados.Tipo.unique()
```

```python
casa = ['Casa', 'Quitinete', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(15)
```
