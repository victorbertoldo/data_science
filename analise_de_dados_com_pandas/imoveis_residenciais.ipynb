## Imóveis residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep=';')
dados.head()
```

### Gerando uma lista contendo apenas os imoveis residenciais
O objetivo aqui é usar a lista como parametro para filtrar o dataframe

```python
residencial = dados['Tipo'].drop_duplicates().to_frame()
residencial = residencial.query('~Tipo.str.contains("comerc|loja|studio|hotel|galp|terr|box|cara|tio|stria|sada|Int", \
                         case=False)', 
                    engine='python') # not like 
        
residencial = list(residencial['Tipo'].values)               
residencial
```

```python
selecao = dados['Tipo'].isin(residencial)
selecao
```

>Utilizando a Series gerada acima como filtro no dataframe, somente serão retornadas as linhas que estiverem como True

```python
dados_residencial = dados[selecao]
dados_residencial.head(10)
```

```python
dados_residencial.shape
```

```python
dados_residencial.index = range(0, dados_residencial.shape[0])
dados_residencial
```

```python
dados_residencial.to_csv('out/aluguel_residencial.csv', sep=';', index=False)
```
