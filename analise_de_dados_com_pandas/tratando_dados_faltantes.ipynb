### Visualizando linhas contendo nulos

```python
dados[dados['Valor'].isnull()]
```

### Vamos eliminar os dados nulos

```python
A = dados.shape[0]
dados.dropna( subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

```python
dados[dados['Valor'].isnull()]
```

### Tratando outros dados faltantes

```python
dados[dados['Condominio'].isnull()].shape[0]
```

### Eliminando do dataframe os dados Tipo apartamento e condominio nulo

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull()) # este é o subset q queremos eliminar
```

```python
A = dados.shape[0]
dados = dados[~selecao] # o '~' inverte a seleção como se fosse um [NOT IN]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

### Todos os valores que correspondiam a condição foram eliminados, porém ainda há valores nulos no valor do condominio
### Estas linhas são importantes portanto iremos preencher com zeros os nulos que ainda restam

```python
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados[dados['IPTU'].isnull()].shape[0]
```

```python
dados.info()
```

> Agora não temos mais dados nulos em nosso dataframe

```python
dados.to_csv('out/aluguel_residencial_notna.csv', sep=';', index= False)
```
